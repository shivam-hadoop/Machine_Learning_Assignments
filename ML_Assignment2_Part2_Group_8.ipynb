<b> <font color='purple'>Importing the Libraries & dataset </font> </b>

```python
from pathlib import Path
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn import preprocessing
import matplotlib.pylab as plt

import dmba

%matplotlib inline
```

```python
university_dataset = pd.read_csv("D:/Monica/IIMA ABA/Modules/Machine Learning/Assignment 2/Universities.csv")
university_dataset.head(5)
```

<b> <font color='purple'>Part A : Removing the Categorical and Missing Data  </font></b>

```python
cleaned_df1 = university_dataset[university_dataset.columns[3:20]]
cleaned_df = cleaned_df1.dropna()
cleaned_df
```

<b> <font color='purple'>Data Summary  </font></b>

```python
pd.DataFrame({'mean': cleaned_df.mean(),
              'sd': cleaned_df.std(),
              'min': cleaned_df.min(),
              'max': cleaned_df.max(),
              'median': cleaned_df.median(),
              'length': len(cleaned_df),
              'miss.val': cleaned_df.isnull().sum(),
             })
```

<b> <font color='purple'>Part B : PCA  </font></b>

```python
pcs = PCA()
pcs.fit(cleaned_df.dropna(axis=0))
pcsSummary_df = pd.DataFrame({'Standard deviation': np.sqrt(pcs.explained_variance_),
                           'Proportion of variance': pcs.explained_variance_ratio_,
                           'Cumulative proportion': np.cumsum(pcs.explained_variance_ratio_)})
pcsSummary_df = pcsSummary_df.transpose()
pcsSummary_df.columns = ['PC{}'.format(i) for i in range(1, len(pcsSummary_df.columns) + 1)]
pcsSummary_df.round(4)
```

<b> <font color='purple'> Without normalising data, we can capture 92% variability with PC1 and PC2. But since the variables are in different units (eg: dollars, percentages, numbers, etc.) it is best to normalise so that they are comparable. Performing PCA on un-normalized variables will lead to insanely large loadings for variables with high variance. In turn, this will lead to dependence of a principal component on the variable with high variance. This is undesirable. </font></b>

<b> <font color='purple'> Normalising data </font></b>

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit_transform(cleaned_df)
y = sc.transform(cleaned_df)

normalised_data = pd.DataFrame(y, columns = cleaned_df.columns)
normalised_data
```

```python
normalised_data.describe().round(2)
```

<b> <font color='purple'> The above data is normalised as the mean = 0  and sd = 1  </font></b>

```python
pcs = PCA()
pcs.fit(normalised_data)
pcsSummary_df = pd.DataFrame({'Standard deviation': np.sqrt(pcs.explained_variance_),
                           'Proportion of variance': pcs.explained_variance_ratio_,
                           'Cumulative proportion': np.cumsum(pcs.explained_variance_ratio_)})
pcsSummary_df = pcsSummary_df.transpose()
pcsSummary_df.columns = ['PC{}'.format(i) for i in range(1, len(pcsSummary_df.columns) + 1)]
pcsSummary_df.round(4)
```

<b> <font color='purple'> After normalising, to capture 90% of the variance in data, we need to include PC1 to PC9. 
    
  There is a massive drop in the variation explananability starting from PC3. For our purpose, we can consider PC1 to PC9 as the key components since they explain huge chunks of variance (more than 90% together) so we would not be losing out on a lot of information. Increasing the principal components increases variability captured but also increases computational effort and may lead to overfitting, hence we would not be considering PC10, PC11 etc. </font></b>

```python
pcsComponents_df = pd.DataFrame(pcs.components_.transpose(), columns=pcsSummary_df.columns, 
                                index=cleaned_df.columns)
pcsComponents_df.sort_values('PC1',ascending=False)
```

<b> <font color='purple'> The first principal component measures the balance between tuition fees (out of state and in state) and the student/faculty ratio. It means that if a university has higher tuition fees, it would have a lower student/faculty ratio </font></b>

```python
pcsComponents_df.sort_values('PC2',ascending=False)
```

<b> <font color='purple'> The second principal component measures the balance between the no of enrolled students (and full time graduates/accepted applicants/received applications) with the tuition fees (out of state and in state). It means that if a university has higher number of enrolled students, its tuition fees would be on the lower side. This is in line with the explanation of PC1 as well. </font> </b>
